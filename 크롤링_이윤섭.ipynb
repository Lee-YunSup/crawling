```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
```

```python
page_no=1
driver=webdriver.Chrome(service=Service(ChromeDriverManager().install()))
page_url="https://search.shopping.naver.com/search/all?frm=NVSCPRO&origQuery=%EC%A0%84%EB%8F%99%ED%82%A5%EB%B3%B4%EB%93%9C&pagingIndex=1&pagingSize=40&productSet=total&query=%EC%A0%84%EB%8F%99%ED%82%A5%EB%B3%B4%EB%93%9C&sort=rel&timestamp=&viewType=list"
title = []
name =[]
price =[]
driver.get(page_url)
SCROLL_PAUSE_SEC = 1

# 스크롤 높이 가져옴
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # 끝까지 스크롤 다운
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # 1초 대기
    time.sleep(SCROLL_PAUSE_SEC)

    # 스크롤 다운 후 스크롤 높이 다시 가져옴
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height


for i in range(1,49):
    a=driver.find_element(By.XPATH,f'//*[@id="content"]/div[1]/div[2]/div/div[{i}]/div/div/div[2]/div[4]/div[1]').get_attribute('innerText')
    b=driver.find_element(By.XPATH,f'//*[@id="content"]/div[1]/div[2]/div/div[{i}]/div/div/div[2]/div[1]/a').get_attribute('innerText')
    c=driver.find_element(By.XPATH,f'//*[@id="content"]/div[1]/div[2]/div/div[{i}]/div/div/div[2]/div[2]').get_attribute('innerText')
    title.append(a)
    name.append(b)
    price.append(c)
    time.sleep(0.3)

```

```python
r_price2=[]
for i in range(0,48):
    if price[i][0]=='광':
        r_price2.append(price[i][2:11])
    elif price[i][0]=='최':
        r_price2.append(price[i][2:11])
    else : r_price2.append(price[i][0:9])
```

```python
r_price3=[]
for i in range(0,48):
    r_price3.append(r_price2[i].strip('\n원판배'))
```

```python
data = pd.DataFrame({"name":name, "price": r_price3, "info": title})
```
